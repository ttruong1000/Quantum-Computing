### **1.1.1 - Bits to Qubits**

- Classical states of computation are either 0 or 1.
- In quantum computing, states can be in superposition; states can be at 0 or 1 at the same time.
- Superpositions allow calculations to be done on states simulataneously - quantum algorithms may have exponential speed up run time.
- Once we measure the superposition state, it collapses into one of its states.
- Designing quantum algorithms uses interference effects

### **1.1.2 - Dirac Notation and Density Matrices**

- Dirac Notation is used to describe quantum states.
- Let $a, b \in \mathbb{C}^2$. Then,
  - ket: $| a \rangle = \begin{pmatrix}
  a_0 \\
  a_1
  \end{pmatrix}$
  - bra: $\langle b | = | b \rangle^* = \begin{pmatrix}
  b_0 \\
  b_1
  \end{pmatrix}^* = \begin{pmatrix} b_0^* & b_1^* \end{pmatrix}$
  - bra-ket: $\langle b | a \rangle = a_0b_0^* + a_1b_1^* = \langle a | b \rangle^*$ 
  - ket-bra: $| a \rangle\langle b | = \begin{pmatrix} a_0b_0^* & a_0b_1^* \\
  a_1b_0^* & a_1b_1^* \end{pmatrix}$
- Orthogonal Pure States
  - $| 0 \rangle = \begin{pmatrix}
  1 \\
  0
  \end{pmatrix}$
  - $| 1 \rangle = \begin{pmatrix}
  0 \\
  1
  \end{pmatrix}$
- Examples
  - $| 0 \rangle\langle 0 | = \begin{pmatrix}
  1 & 0
  \end{pmatrix}\begin{pmatrix}
  1 \\
  0
  \end{pmatrix} = \begin{pmatrix}
  1 & 0 \\
  0 & 0
  \end{pmatrix}$
  - $| 1 \rangle\langle 1 | = \begin{pmatrix}
  0 & 1
  \end{pmatrix}\begin{pmatrix}
  0 \\
  1
  \end{pmatrix} = \begin{pmatrix}
  0 & 0 \\
  0 & 1
  \end{pmatrix}$
  - $\rho = \begin{pmatrix}
  \rho_{00} & \rho_{01} \\
  \rho_{10} & \rho_{11}
  \end{pmatrix} = \rho_{00} | 0 \rangle\langle 0 | + \rho_{01} | 0 \rangle\langle 1 | + \rho_{10} | 1 \rangle\langle 0 | + \rho_{11} | 1 \rangle\langle 1 |$

  \rho_{10} & \rho_{11}
